## 1. Figures and Subplots - Not as cool as Matlab

```python
# plots in this library exist in a figure object. A figure can be created as follows
fig = plt.figure()

# we can not make a plot with an empty figure. this can be rather used to make subplots

## this command will make a subplot among 4 at position 1
ax1 = fig.add_subplot(2,2,1)
# the following will create subplots at the corresponding positions
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,2,3)

# after adding this, we get a plot at the last subplot created 

plt.plot(np.random.randn(50).cumsum(), 'k--')
```

the 'k--' is a style option instructing matplotlib to plot a black dashed line. 
the objects returned by fig.add_subplot here are AxesSubplot object son which we can directly plot on the other empty subplots by calling each one's instance method.

```python
_  =  ax1.hist(np.random.randn(100), bins=20, color='k', alpha=0.3)
ax2.scatter(np.arange(30), np.arange(30) + 3 * np.random.randn(30))
```

```python
# creating a figure with a grid of subplots is a very common tasks (even in matlab), so matplotlib includes a convenience method, plt.subplots.
# this method creates a new figure and returns a numpy array containing the created subplot objects:

fig, axes = plt.subplots(2,3)
axes
```

### Adjusting spacing around subplots
Spacing around the outside of subplots and the spacing between plots ca be changed using the subplots_adjust method on figure objects. 

subplots_adjust(left=None, bottom=None, right=None, top=None, wspace=None, hspace=None)

wspace and hspace control the percent of the figure width and height, respectively to use as spacing between subplots.

```python
fig, axes = plt.subplots(2,2, sharex=True, sharey=True)
for i in range(2):
    for j in range(2):
        axes[i,j].hist(np.random.randn(500), bins=50, color='k', alpha=0.5)
        
# originally, we have this
```

```python
fig, axes = plt.subplots(2,2, sharex=True, sharey=True)
for i in range(2):
    for j in range(2):
        axes[i,j].hist(np.random.randn(500), bins=50, color='k', alpha=0.5)

        
plt.subplots_adjust(wspace=0, hspace=0)
# on this added, we get this
```

## 2. Colors, Markers and Line Styles
The plot function of matplotlib accepts arrays of x and y coordinates and optionally a string abbreviation indicating the color and line style for a plot.

```python
x = np.random.randn(25)
y = np.random.randn(25)
```

```python
# this reps a plot of x against y with green dashes
plt.plot(x, y, 'g--')
```

```python
# this same plot could be created using the following:
plt.plot(x,y, linestyle='--', color='g')
```

```python
# another method of doing this, is by using color hexes
plt.plot(x,y, linestyle='-', color='#DA3132')
```

```python
# line plots can additionally have markers to highlight the actual data points.
# the marker can be part of the style string, which must have olor followed by the type and line style
z = np.random.randn(25).cumsum()
plt.plot(z, 'ko--')
```

```python
# the above could alsp be written as
plt.plot(z, color='k', linestyle='dashed', marker='o')
```

For line plots, we notice that subsequent points are linearly interpolated by default. This can be altered with the drawstyle option.

```python
data = np.random.randn(25).cumsum()
data
```

```python
plt.plot(data, 'k--', label='Default')
```

```python
plt.plot(data, 'k-', drawstyle='steps-post', label='steps-post')
```

```python
# combining the previous two, we get
plt.plot(data, 'k--', label='Default')
plt.plot(data, 'k-', drawstyle='steps-post', label='steps-post')
plt.legend(loc='best')
```

## 3. Ticks, Labels, Legends
For most kinds of plot decorations, there are two main ways to do things: using the procedural pyplot interface (matplotlib.pyplot) and the more object-oriented native matplotlib API.

### Setting title, axis, labels, ticks and ticklabels

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
z = np.random.randn(1000).cumsum()

# to change, the x-axis ticks, it is easiest to use the set_xticks and set_xticklabels.
ticks = ax.set_xticks([0, 250, 500, 750, 1000])
labels = ax.set_xticklabels(['one', 'two', 'three', 'four', 'five'],
                           rotation=30, fontsize='small')
ax.set_xlabel('Stages')
ax.set_title('My something matplotlib plot')
ax.plot(z)
```

### Adding legends
Legends are a very important part of plots for identifying playing elements. 

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
x = np.random.randn(500).cumsum()
ax.plot(x, 'k', label='one')

y = np.random.randn(500).cumsum()
ax.plot(y, 'k--', label='two')

z = np.random.randn(500).cumsum()
ax.plot(z, 'k.', label='three')

ax.legend(loc='best')
```

## 4. Annotations and Drawing on a Subplot

```python
# text
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
x = np.random.randn(500).cumsum()
ax.plot(x, 'k', label='one')

y = np.random.randn(500).cumsum()
ax.plot(y, 'k--', label='two')

z = np.random.randn(500).cumsum()
ax.plot(z, 'k.', label='three')

ax.text(200, 10, 'Hello Annotation', family='monospace', fontsize=10)

ax.legend(loc='best')
```

```python
# drawing
fig = plt.figure()
ax = fig.add_subplot(1,1,1)

rect = plt.Rectangle((0.2, 0.75), 0.4, 0.15, color='k', alpha=0.3)
circ = plt.Circle((0.7, 0.2), 0.15, color='b', alpha=0.3)
poly = plt.Polygon([[0.15, 0.15], [0.35, 0.4], [0.2, 0.6]], color='g', alpha=0.3)

ax.add_patch(rect)
ax.add_patch(circ)
ax.add_patch(poly)
```

## 5. Saving Plots to File
You can save the active figure to file using plt.savefig.

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
x = np.random.randn(500).cumsum()
ax.plot(x, 'k', label='one')

y = np.random.randn(500).cumsum()
ax.plot(y, 'k--', label='two')

z = np.random.randn(500).cumsum()
ax.plot(z, 'k.', label='three')

ax.text(200, 10, 'Hello Annotation', family='monospace', fontsize=10)

ax.legend(loc='best')
plt.savefig('figexample.svg')
```

The file type is inferred from the extension we provide.
We can add dpi(dots per inch) and bbox_inches keywords to can trim the whitespace around the actual figure.


```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
x = np.random.randn(500).cumsum()
ax.plot(x, 'k', label='one')

y = np.random.randn(500).cumsum()
ax.plot(y, 'k--', label='two')

z = np.random.randn(500).cumsum()
ax.plot(z, 'k.', label='three')

ax.text(200, 10, 'Hello Annotation', family='monospace', fontsize=10)

ax.legend(loc='best')
plt.savefig('figexample2.png', dpi=400, bbox_inches='tight')
```
