# Line Charts
As we have already seen, we can make line charts using plt.plot().

```python
# imports
import matplotlib.pyplot as plt
```

```python
#vars
variance = [1,2,4,8,16,32,64,128,256]
bias_squared = [_ for _ in reversed(variance)]
total_error = [x+y for (x,y) in zip(variance,bias_squared)]
xs = [ i for i,_ in enumerate(variance)]
```

```python
# we can make multiple calls to plt.plot()
# to show multiple series on the same chart
plt.plot(xs, variance, 'g-', label="variance")
plt.plot(xs, bias_squared, 'b-.', label="bias^2")
plt.plot(xs, total_error, 'r:', label="total error")

# because we have assigned labels to each series
# we can get a legend a free legend
# loc=9 means top center
plt.legend(loc=9)
plt.xlabel("model complexity")
plt.title("The Bias-Variance Tradeoff")
plt.savefig("/home/kanye/Development/JupyterLab/PythonDS/3-DataVisualisation/3-linecharts-output.jpg")
plt.show()
```
