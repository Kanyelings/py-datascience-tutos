For this, we shall use the matplotlib.pyplot module. 
This module maintains an internal state which you build up visualisation step by step. Once you are done, you can save ut using 'savefig()' or display it with 'show()'.

```python
# importing the module from matplotlib
from matplotlib import pyplot as plt
```

```python
# initialising vars
years = [1950, 1960, 1970, 1980, 1990, 2000, 2010]
gdp = [300.2, 543.3, 1075.9, 2862.5, 5979.6, 10289.7, 14958.3]
```

```python
# add a title
plt.title("Nominal GDP")
```

```python
# add a label to the y-axis
plt.ylabel("Billions of $")
```

```python
# plotting the years on the x-axis and gdp on y-axis
plt.plot(years, gdp)
```

```python
# plotting with some other params
plt.plot(years, gdp, color='green', marker='x', linestyle='solid')
```
